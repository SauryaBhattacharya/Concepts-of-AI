Task - 3 - Outlier Detection:

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "/content/drive/MyDrive/Assignment-1(AI)/south_asian_countries.csv"
south_asia_df = pd.read_csv(path)

#1 Identify outlier countries in South Asia based on their Score and GDP per Capita.
#2 Define outliers using the 1.5  IQR rule.
def find_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]
score_outliers = find_outliers(south_asia_df, 'score')
gdp_outliers = find_outliers(south_asia_df, 'Log GDP per capita')

outliers = pd.concat([score_outliers, gdp_outliers]).drop_duplicates()

#3 Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
plt.figure(figsize=(10, 6))
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], label="Non-Outliers", color='blue')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label="Outliers", s=100)
for _, row in south_asia_df.iterrows():
    plt.annotate(row['Country name'], (row['Log GDP per capita'], row['score']),
                 textcoords="offset points", xytext=(5, 5), ha='center', fontsize=8, color='darkred')

plt.xlabel("GDP per Capita (Log)")
plt.ylabel("Score")
plt.title("Scatter Plot: GDP per Capita vs Score (Outliers Highlighted)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
print("Outlier Countries:")
print(outliers[['Country name', 'Log GDP per capita', 'score']])

```

Outliers Characteristics:

Outliers in South Asia, such as Afghanistan, display unique characteristics, including extremely low GDP per capita and Happiness Scores, or high discrepancies between economic indicators and well-being metrics. Countries like Afghanistan struggle with economic challenges and low health outcomes, leading to low scores. These factors make Afganistan stand out as statistical outliers, reflecting significant social and economic variability within the region.

Impact of Outliers on Regional Averages:

The impact of these outliers on regional averages is profound, as they skew the mean and distort the overall perception of South Asias economic and social conditions. Countries with relatively better performance, like Sri Lanka or India, may be overshadowed by the low regional average, masking their positive trends. This can lead to policy misalignment, as focusing solely on the average may overlook the unique challenges faced by outlier countries, necessitating tailored interventions for them.

Task - 4 - Exploring Trends Across Metrics:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = "/content/drive/MyDrive/Assignment-1(AI)/south_asian_countries.csv"
south_asia_df = pd.read_csv(path)

#1 Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation (pearson correlation) with the Score for South Asian countries.
metrics = ['Freedom to make life choices', 'Generosity']
correlations = {}
for metric in metrics:
    correlations[metric] = south_asia_df['score'].corr(south_asia_df[metric], method='pearson')

print("Pearson correlations with Score: ")
for metric, corr in correlations.items():
    print(f"{metric}: {corr:.3f}")

#2 Create scatter plots with trendlines for these metrics against the Score (using matplotlib)
for metric in metrics:
    plt.figure(figsize=(10, 6))
    plt.scatter(south_asia_df[metric], south_asia_df['score'], color='blue', label='Data')
    p = np.polyfit(south_asia_df[metric], south_asia_df['score'], 1)
    plt.plot(south_asia_df[metric], np.polyval(p, south_asia_df[metric]), color='red', label='Trendline')
    plt.xlabel(metric)
    plt.ylabel("Score")
    plt.title(f"Scatter Plot: {metric} vs Score")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.show()

#3 Identify and discuss the strongest and weakest relationships between these metrics and the Score for South Asian countries.
strongest_metric = max(correlations, key=correlations.get)
weakest_metric = min(correlations, key=correlations.get)

print("\nDiscussion:")
print(f"The strongest relationship is between Score and {strongest_metric} with a correlation of {correlations[strongest_metric]:.3f}.")
print("This indicates that as Generosity increases in South Asian countries, \nthere is a notable rise in the overall Score (e.g., happiness or well-being). \nGenerosity plays a significant role in contributing to a positive societal environment.")
print(f"\nThe weakest relationship is between Score and {weakest_metric} with a correlation of {correlations[weakest_metric]:.3f}.")
print("While still positive, the slightly lower correlation suggests that Freedom \nto Make Life Choices has a less significant influence on the Score compared \nto Generosity. This indicates that other factors may play a more prominent role in determining overall well-being.")
```

Task - 5 - Gap Analysis:

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "/content/drive/MyDrive/Assignment-1(AI)/south_asian_countries.csv"
south_asia_df = pd.read_csv(path)

#1 Add a new column, GDP-Score Gap, which is the difference between GDP per Capita and the Score for each South Asian country.
south_asia_df['GDP-Score Gap'] = south_asia_df['Log GDP per capita'] - south_asia_df['score']

#2 Rank the South Asian countries by this gap in both ascending and descending order.
ascending_gap = south_asia_df.sort_values(by='GDP-Score Gap', ascending=True)
descending_gap = south_asia_df.sort_values(by='GDP-Score Gap', ascending=False)

#3 Highlight the top 3 countries with the largest positive and negative gaps using a bar chart.
top_positive_gap = descending_gap.head(3)
top_negative_gap = ascending_gap.head(3)
top_gaps = pd.concat([top_positive_gap, top_negative_gap]) #for visualization
plt.figure(figsize=(10, 6))
plt.bar(top_gaps['Country name'], top_gaps['GDP-Score Gap'], color=['green' if x > 0 else 'darkblue' for x in top_gaps['GDP-Score Gap']])
plt.axhline(0, color='black', linewidth=0.8)
plt.xlabel('Country Name')
plt.ylabel('GDP-Score Gap')
plt.title('Top 3 Positive and Negative GDP-Score Gaps in South Asian Countries')
plt.xticks(rotation=45)
plt.show()
```

4. Discussion:

The GDP-Score Gap highlights a mismatch between economic performance (GDP per capita) and perceived well-being (Score). Countries like Nepal, Pakistan, and India show the largest negative gaps, indicating that economic prosperity in these nations does not necessarily translate into higher happiness scores.

Reasons behind the gaps:

Unequal Distribution of Wealth:
High GDP per capita may not benefit the entire population due to economic inequality. Countries like India and Pakistan face significant wealth disparity, which suppresses happiness scores despite rising GDP.

Social and Political Challenges:
Political instability, corruption, and governance issues often prevent people from fully benefiting from economic growth. For example, Nepal has experienced historical political turmoil, contributing to its large negative gap.

Quality of Life Factors:
GDP per capita measures economic output but overlooks key social factors like healthcare, education, and work-life balance. Countries with lower investments in these areas tend to have lower well-being scores.








Implications for south asian countries

To bridge the GDP-Score gap, South Asian countries need to prioritize inclusive economic growth by ensuring equitable wealth distribution so that economic gains translate into improved well-being. Governments must address pressing social issues such as political instability, limited healthcare access, and inadequate education systems, which often hinder progress. A strong focus on human developmentthrough investments in health, education, and freedombeyond GDP is essential to enhance happiness scores. Furthermore, fostering regional collaboration among South Asian nations can help tackle shared challenges and drive economic and social reforms, ultimately improving overall regional well-being.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "/content/drive/MyDrive/Assignment-1(AI)/WHR-2024-5CS037.csv"
df = pd.read_csv(file_path)

# Identify outlier countries in both regions based on Score and Log GDP per Capita.
def identify_outliers(data, column_name):
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    return outliers

south_asia_score_outliers = identify_outliers(south_asia_df, 'score')
south_asia_gdp_outliers = identify_outliers(south_asia_df, 'Log GDP per capita')
middle_east_score_outliers = identify_outliers(middle_east_df, 'score')
middle_east_gdp_outliers = identify_outliers(middle_east_df, 'Log GDP per capita')
south_asia_outliers = pd.concat([south_asia_score_outliers, south_asia_gdp_outliers]).drop_duplicates()
middle_east_outliers = pd.concat([middle_east_score_outliers, middle_east_gdp_outliers]).drop_duplicates()

print("South Asia Outliers based on Score and Log GDP per Capita:")
print(south_asia_outliers[['Country name', 'score', 'Log GDP per capita']])
print("\nMiddle East Outliers based on Score and Log GDP per Capita:")
print(middle_east_outliers[['Country name', 'score', 'Log GDP per capita']])

# Plot these outliers and discuss their implications.
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# South Asia
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, ax=axes[0], color='green', label='Data')
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_outliers, ax=axes[0], color='red', label='Outliers')
axes[0].scatter(south_asia_outliers['Log GDP per capita'], south_asia_outliers['score'], color='red', label='Outliers')
axes[0].set_title('South Asia: Outliers based on Log GDP per Capita and Score')
axes[0].set_xlabel('Log GDP per Capita')
axes[0].set_ylabel('Score')
axes[0].grid(True)
axes[0].legend()

# Middle East
axes[1].scatter(middle_east_df['Log GDP per capita'], middle_east_df['score'], color='blue', label='Data')
axes[1].scatter(middle_east_outliers['Log GDP per capita'], middle_east_outliers['score'], color='red', label='Outliers')
axes[1].set_title('Middle East: Outliers based on Log GDP per Capita and Score')
axes[1].set_xlabel('Log GDP per Capita')
axes[1].set_ylabel('Score')
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```

This suggests that Afghanistan's data significantly deviates from the rest of the countries in the region. The outlier status could be attributed to factors such as political instability, ongoing conflict, or socio-economic challenges unique to Afghanistan, which may affect its development metrics. This anomaly warrants further exploration into Afghanistan's economic and social conditions, and could influence policy decisions or economic modeling. Excluding or specifically accounting for Afghanistan in data analyses may be necessary to ensure that its outlier status doesn't distort regional trends.

In the Middle East, there are no outliers, indicating a more consistent relationship between Log GDP per capita and score across the region. The absence of outliers suggests that the countries in the Middle East follow a more uniform developmental pattern, possibly due to shared economic structures, social systems, or political stability.
