```python
#Problem-3
import numpy as np

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

#1
add_result = x+y
print("Addition of x and y:\n",add_result)

#2
subtract_result = x - y
print("\nSubtraction of x and y:\n", subtract_result)

#3
multiply_result = x * 2
print("\nMultiplying x by 2:\n", multiply_result)

#4
square_result = np.square(x)
print("\nSquare of each element of x:\n", square_result)

#5
dot_v_w = np.dot(v, w)
print("\nDot product of v and w:", dot_v_w)
dot_x_v = np.dot(x, v)
print("\nDot product of x and v:", dot_x_v)
dot_x_y = np.dot(x, y)
print("\nDot product of x and y:\n", dot_x_y)

#6
concat_x_y = np.concatenate((x, y), axis=0)
print("\nConcatenate x and y along row:\n", concat_x_y)
concat_v_w = np.concatenate((v.reshape(-1, 1), w.reshape(-1, 1)), axis=1)
print("\nConcatenate v and w along column:\n", concat_v_w)

```
